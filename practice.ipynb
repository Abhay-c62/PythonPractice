```python
#Python - Bitwise Operators : &,|,^,~,<<,>>
a=10
b=6
print("a&b:",a&b)
print("a|b:",a|b)
print("a^b:",a^b)
print("~a:",~a)
print("a<<2:",a<<2)
print("a>>2:",a>>2)
```

```python
# Left shift operator (<<) shifts the bits of the number to the left by the specified number of positions. Each left shift operation effectively multiplies the number by 2. For example, if you left shift the binary representation of 10 (which is 1010) by 2 positions, you get 101000, which is equal to 40 in decimal.
a=4
print(a<<1)
print(a<<2)
```

```python
# Right shift operator (>>) shifts the bits of the number to the right by the specified number of positions. Each right shift operation effectively divides the number by 2. For example, if you right shift the binary representation of 10 (which is 1010) by 2 positions, you get 0010, which is equal to 2 in decimal.
a=5
print(a>>1)
print(a>>2)
print(a>>3)
print(a>>4)
```

```python
#Use a while loop to compute the factorial of a number provided by the user.
num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"The factorial of {num} is {factorial}")
```

```python
#WAP use a list Compression to create a list of all vowels found in a given sentence
#Step-1:take user input
sentence = input("Enter a sentence: ")
#Step-2: define the vowels both upper and lower case
vowels = "AEIOUaeiou"
#Step-3: use list comprehension to filter vowels from the sentence
filter_vowels = [char for char in sentence if char in vowels]
#Step-4: print the result
print(f"The vowels in the sentence are: {filter_vowels}")
```

```python
#parameters: Default,positional,keyword,*args,**kwargs
def describe_pet(name, animal="dog"):
    """Display information about a pet."""
    print(f"{name} is a {animal}.")
describe_pet("Rocky")  # Uses default animal type "dog"
describe_pet("Ronny", "cat")  # Overrides default animal type with "cat

def add_all(*args): #Variable positional args (tuple)
    """Return the sum of all arguments."""
    return sum(args)
print(add_all(1, 2, 3, 4)) 

def print_info(**kwargs): #Variable keyword args (using dictionary)
    """Print key-value pairs from keyword arguments."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=23, city="New York")
```

```python
# Return value (without explicit and implicitly)
def min_max(numbers):
    return min(numbers), max(numbers)
low, high = min_max([5,11,3,8,2])
print(f"Min: {low}, Max: {high}")
```
